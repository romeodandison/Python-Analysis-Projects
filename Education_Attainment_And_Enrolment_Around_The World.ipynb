Columns description : 

Age[15-49]All_gradeX : Percentage of the entire population ages 15 to 49 that has completed grade X
Age[15-49]Male_gradeX : Percentage of the male population ages 15 to 49 that has completed grade X
Age[15-49]Female_gradeX : Percentage of the female population ages 15 to 49 that has completed grade X
X = [1...3]

```python
#Import Necessary Library

import pandas as pd

#Import the dataset

dataset = pd.read_csv('EDUCATION_ATTAINMENT.csv')
```

```python
#Display general information about the dataset

dataset.info()
```

```python
#Descriptive analysis on the dataset using the 'describe' function

dataset.describe()
```

```python
#Data exploration with ydata-profiling

from ydata_profiling import ProfileReport

# Generate the profile report
edu_report = ProfileReport(dataset, title = 'Education Attainment And Enrolment Around The World')
```

```python
# Display the report
edu_report.to_notebook_iframe()
```

```python
# Or generate an HTML report
edu_report.to_file("edu_attainment.html")
```
